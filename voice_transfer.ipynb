### sound treatment

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import IPython
import soundfile as sf
```

```python
y, sr = librosa.load("./gura-[AudioTrimmer.com].mp3", sr=None)
```

```python
n_fft=2048
ps = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft)
ps_db= librosa.power_to_db(ps, ref=np.max)
librosa.display.specshow(ps_db, x_axis='s', y_axis='log')
```

```python
res = librosa.feature.inverse.mel_to_audio(ps, sr, n_fft= n_fft)
```

```python
sf.write("file_trim_5s.wav", res, sr, subtype='PCM_24')
```
